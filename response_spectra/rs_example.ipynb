```python
import os
import re
import numpy as np
import matplotlib.pyplot as plt
from response_spectra import get_response_spectrum
```

```python
# load earthquake data
acc = np.loadtxt('EEE1.dat')

# earthquake time step
eq_dt = 0.01

# frequency range
freq = np.arange(0.01, 60.1, 0.1)

# damping ratio
xi = 0.05

# spectral acceleration
Sa = get_response_spectrum(acc, xi, eq_dt, 1/freq)
```

```python
plt.rcParams['figure.dpi'] = 300

# settings for jupyter book: svg or pdf
%matplotlib inline
import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg', 'pdf')
os.makedirs('out', exist_ok=True)
```

```python
plt.style.use('ppt14grid.mplstyle')

plt.figure(figsize=(6, 3.75))
plt.plot(freq, Sa)
plt.grid(which='minor', linewidth=0.15, alpha=0.35)
plt.xscale('log')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Spectral Acceleration (g)')
plt.savefig('out/rs.pdf')
plt.savefig('out/rs.svg')
plt.show()
```
